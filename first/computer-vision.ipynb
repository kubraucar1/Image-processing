```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv.imread('picture.jpg')   # Read the image
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
cv.imshow("image",img_rgb)
key = cv.waitKey(0) 
if key == 27:
    cv.destroyAllWindows()
#to close, press ESC
```

```python
print("Image type: ", type(img_rgb))
print("Image data type: ", img_rgb.dtype)
print("Image shape: ", img_rgb.shape)
```

```python
line_img = img_rgb.copy()        # Make a copy of the image
cv.line(line_img,(0,0),(250,380),(98,47,89),5)   # Draw the line
plt.imshow(line_img)    # show the image
```

```python
rectangle_img = img_rgb.copy()
cv.rectangle(rectangle_img,(70,50),(200,160),(0,0,0),3)
plt.imshow(rectangle_img)
'''
img_rgb, dikdörtgenimizi çizmek istediğimiz resim
(70,50), pt1, dikdörtgenin köşesi
(200,160), pt2, pt1'in dikdörtgeninin dikey metni
(0,0,0)circle_img = img_rgb.copy() 
cv.circle(circle_img,(100,130), 90, (0,0,255), -1)
plt.imshow(circle_img)), dikdörtgen rengimizin RGB değeri
3 , dikdörtgenimizin piksel cinsinden kalınlığı
'''
```

```python
circle_img = img_rgb.copy() 
cv.circle(circle_img,(125,225), 50, (0,0,0), -1)
plt.imshow(circle_img)
'''
img_rgb, dikdörtgenimizi çizmek istediğimiz resim
(125,225), dairenin başladığı yerin merkezi konumu
50, dairenin piksel cinsinden yarıçapı
(0,0,0), dikdörtgen rengimizin RGB değeri, bu durumda mavidir
-1 dairemizin kalınlığı, NEGATİF ise, o zaman daire doldurulacaktır.
'''
```

```python
text_img = img_rgb.copy()                
font = cv.FONT_HERSHEY_SIMPLEX 
textcolor = (255,255,255)
cv.putText(text_img,'Mustafa Kemal Ataturk <3',(0,30),font, 1,textcolor,2,cv.LINE_AA)             
plt.imshow(text_img) 
```

```python
cv.imwrite("Atatürk_with_Text.jpg",text_img)
```

```python
image = cv.imread('picture.jpg', 1)
# Loading the image
 
half = cv.resize(image, (0, 0), fx = 0.5, fy = 0.5)
bigger = cv.resize(image, (1050, 1610))
 
stretch_near = cv.resize(image, (780, 540), 
               interpolation = cv.INTER_LINEAR)
 
 
Titles =["Original", "Half", "Bigger", "Interpolation Nearest"]
images =[image, half, bigger, stretch_near]
count = 4
 
for i in range(count):
    plt.subplot(2, 2, i + 1)
    plt.title(Titles[i])
    plt.imshow(images[i])
 
plt.show()
```
